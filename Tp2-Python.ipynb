```python
# Exercice 1

def chercher(T, nombre):
    for i, val in enumerate(T):
        if val == nombre:
            return i
    return -1

def calculermoyenne(T):
    return sum(T)/len(T) if T else 0

def inverser(T):
    return T[::-1]

def supprimer_occurrences(T, x):
    return [val for val in T if val != x]

```

```python
# Exercice 2

def crible_eratosthene(n):
    premiers = []
    sieve = [True]*n
    for p in range(2, n):
        if sieve[p]:
            premiers.append(p)
            for i in range(p*2, n, p):
                sieve[i] = False
    return premiers
```

```python
# Exercice 3

def derive_polynome(P):
    return [coeff*(len(P)-i-1) for i, coeff in enumerate(P[:-1])]

def multiplier_polynomes(P1, P2):
    deg1, deg2 = len(P1)-1, len(P2)-1
    res = [0]*(deg1+deg2+1)
    for i, a in enumerate(P1):
        for j, b in enumerate(P2):
            res[i+j] += a*b
    return res
```

```python
# Exercice 4

def MedianeT(T):
    T_sorted = sorted(T)
    n = len(T_sorted)
    if n % 2 == 1:
        return T_sorted[n//2]
    else:
        return (T_sorted[n//2-1] + T_sorted[n//2])/2
```

```python
# Exercice 5

def AfficheSousSequencesT(T):
    sous_sequences = []
    ss = [T[0]]
    for i in range(1, len(T)):
        if T[i] > T[i-1]:
            ss.append(T[i])
        else:
            if len(ss) > 1:
                sous_sequences.append(ss)
            ss = [T[i]]
    if len(ss) > 1:
        sous_sequences.append(ss)
    return sous_sequences
```

```python
# Exercice 6

def IntersectionT1_T2(T1, T2):
    return list(set(T1) & set(T2))

def ComplementT1(T1, N):
    return [x for x in range(N+1) if x not in T1]

def FusionT1_T2(T1, T2):
    return sorted(list(set(T1) | set(T2)))
```

```python
# Exercice 7

def longueur_nom_sans_espaces(nom):
    return len(nom.replace(' ', ''))
```

```python
# Exercice 8

def nb_occurrences(chaine, x):
    return chaine.count(x)

def inverser_chaine_rec(chaine):
    if len(chaine) <= 1:
        return chaine
    return inverser_chaine_rec(chaine[1:]) + chaine[0]

def est_palindrome_rec(chaine):
    chaine_nettoyee = ''.join([c.lower() for c in chaine if c.isalpha()])
    return chaine_nettoyee == chaine_nettoyee[::-1]
```

```python
# Exercice 10

def comparer_lexicographique(ch1, ch2):
    if ch1 < ch2:
        return f"{ch1} précède {ch2}"
    elif ch1 > ch2:
        return f"{ch1} suit {ch2}"
    else:
        return "Les deux chaînes sont identiques"
```

```python
# Exercice 11

def conjuguer_er(verbe):
    if not verbe.endswith('er'):
        return "Ce n'est pas un verbe en -er."
    radical = verbe[:-2]
    return {
        "je": radical + "e",
        "tu": radical + "es",
        "il": radical + "e",
        "nous": radical + "ons",
        "vous": radical + "ez",
        "ils": radical + "ent",
    }
```
