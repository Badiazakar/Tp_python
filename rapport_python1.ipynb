## EX7 :

```python
def suite(n):
    U0 = 1
    U1 = 2

    if n == 0:
        return U0
    elif n == 1:
        return U1

    for i in range(2, n + 1):
        Un = U1 + 6 * U0
        U0 = U1
        U1 = Un

    return U1
```

```python
def dichotomie(f, a, b, epsilon):
    while (b - a) > epsilon:
        m = (a + b) / 2
        if f(a) * f(m) < 0:
            b = m
        else:
            a = m
    return (a + b) / 2
def f(x):
    return x**2 - 2
```

```python
dichotomie(f, 2, 12, 0.001)
```

## EX11:

```python
def approx(x, n):
    s = 1          
    P = 1  
    F = 1       

    i = 1
    while i <= n:
        P = P * x
        F = F * i
        s = s + P / F
        i = i + 1
    return s
```

```python
approx(3,4)
```

## EX12:

```python

def approximate_integral(a, b, n):
    import math
    dx = (b - a) / n
    integral = 0.0
    integral += 0.5 * math.exp(-a**2)
    integral += 0.5 * math.exp(-b**2)
    for i in range(1, n):
        x = a + i * dx
        integral += math.exp(-x**2)
    integral *= dx
    return integral
```

```python
approximate_integral(6, 9, 2)
```

```python
def plus_grand_diviseur(n):
    p = n - 1
    while p > 1:
        if n % p == 0:
            return p
        p = p - 1
    return -1
```

## EX15 :

```python
def premier(x):
    if x < 2:
        return False
    i = 2
    while i < x:
        if x % i == 0:
            return False
        i = i + 1
    return True

def PlusGranDivPrem(n):
    if n == 1:
        return 1
    p = n
    while p > 1:
        if n % p == 0 and premier(p):
            return p
        p = p - 1
    return 1
```

```python
PlusGranDivPrem(100)
```

## EX16 :

```python
def Factoriser(n):
    facteurs = []
    i = 2
    while n > 1:
        while n % i == 0:
            facteurs.append(i)
            n //= i
        i += 1
    return '*'.join(map(str, facteurs))
```

```python
Factoriser(60)
```

## EX17 :

```python
def nbr_mechant(n):
    if n == 1:
        return True
    for p in [2, 3, 5]:
        while n % p == 0:
            n //= p
    return n == 1

def verifier_mechant(n):
    if nbr_mechant(n):
        print("oui c'est un nombre méchant")
    else:
        print("non, ce n'est pas nombre méchant")
```
